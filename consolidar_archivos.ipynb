# UNIFICACION DE ARCHIVOS EN UN SOLO .CSV
---


Se descargaron los 16 archivos .csv del sistema interno de la Universidad, el cual contiene la información de las inscripciones de cada SEMESTRE del programa de Geología, desde el 2016 hasta el ultimo del 2023.

Lo que pretendemos en este collab, es leer todos los archivos y unificarlos en 1 solo csv para cargarlo en GIT HUB.

* No se realizó ningun procesamiento de data.

Luego de tenerlo en GIT HUB, lo usaremos para desarrollar el proyecto del curso



```python
from google.colab import drive
import os
drive.mount('/content/gdrive')
import pandas as pd
import numpy as np
```

```python
%cd '/content/gdrive/MyDrive/DATA_INSCRITOS'
#Leemos todos los archivos del directorio y los almacenamos en una lista llamada "files"
files=os.listdir('/content/gdrive/MyDrive/DATA_INSCRITOS')
print(files)
```

```python
#Creamos una lista que contenga los dataframes obtenidos de cada CSV, y luego concatenamos los elementos de esa lista en un solo DF
data=[]
for i in range (len(files)):
  data.append(('df' +  str(i)))
  data[i] = pd.read_csv(files[i], sep=';')
  df=pd.concat(data,axis=0, ignore_index= True)
df.to_csv('data_inscripciones.csv', sep=';', encoding='UTF-8')
#Exportamos a un CSV que será cargado en GIT HUB, a partir del cual, se desarrollará el proyecto.
```
